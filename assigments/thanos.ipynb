```python
import os

def thanos_folder(folder_path):
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
        return

    for i in range(1, 5):
        with open(os.path.join(folder_path, f"file_{i}.txt"), 'w') as f:
            f.write("nardien")


    #list of all files currently in the folder
    files = os.listdir(folder_path)
    
    half_count = len(files) // 2
    
    for i in range(half_count):
        file_path = os.path.join(folder_path, files[i])
        os.remove(file_path)

path = "folder_for_files"
thanos_folder(path)
```
