```python
email = "Amit_ml@gmail.edu"

if "@" in email and "." in email:
    # Split into: ['Amit_ml', 'gmail.edu']
    user, full_domain = email.split("@")
    # Split domain into: ['gmail', 'edu']
    domain_name, extension = full_domain.split(".")

    print(f"User: {user}")
    print(f"Domain: {domain_name}")

    # Simplified mapping
    types = {"com": "commercial", "edu": "educational"}
    print(f"{types.get(extension, 'other')} domain")
else:
    print("Invalid email")
```

```python
def clean_message(msg, swap_rules):
    parts = msg.strip("&*$!@#0123456789").split()
    
    first = parts[0][::-1]
    
    second = parts[1]
    for old, new in swap_rules.items():
        second = second.replace(old, new)
    
    return f"{first} {second}"





# Task 3
print(clean_message("&&&**$gnirtS PLIO!!@1234", {"I": "E", "O": "U"}))
# Task 4
print(clean_message("##$$$@!yalpstcejorp EPUVT****9887", {"E": "A", "U": "O"}))
```
